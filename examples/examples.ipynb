# See how many symbols per category

```python
import binance_syncer as bs

for market_type, value in bs.SCHEMA.items():
    print(f"Market Type: {market_type}")
    for data_type in value:
        print(f"  Data Type: {data_type}")

        # Configure syncer
        syncer = bs.Syncer(
            market_type,
            data_type,
            bs.KlineInterval.D1,
            s3=True
        )

        symbols = await syncer.list_remote_symbols()
        print(f"    Available Symbols: {len(symbols)}")
```

# Sync first symbol of every category (Can take a while for "TRADE" categories)

```python
import binance_syncer as bs

for market_type, value in bs.SCHEMA.items():
    print(f"Market Type: {market_type}")
    for data_type in value:
        print(f"  Data Type: {data_type}")

        # Configure syncer
        syncer = bs.Syncer(
            market_type,
            data_type,
            bs.KlineInterval.D1,
        )

        symbols = await syncer.list_remote_symbols()

        symbol = symbols[:1]
        print(f"    Syncing: {symbol[0]}")

        await syncer.sync(symbol)
```
